取國道3號的其中一個即時攝影機作為範例

```python
root = highway_3['html'][655]
```

擷取網路影像存取10秒影像，並將影片儲存成mp4檔

```python
cap = cv2.VideoCapture(root)
ret, frame = cap.read()
height, weight, channel = frame.shape
fourcc = cv2.VideoWriter_fourcc(*'MP4V')
out = cv2.VideoWriter('output.mp4', fourcc, 20, (weight, height))

while (True):
    ret, frame = cap.read()
    if ret:
        out.write(frame)
        cv2.imshow('frame', frame)
        key = cv2.waitKey(90)
        if   key == 27:
            break
    else:
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
